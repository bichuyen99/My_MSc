```python
import numpy as np
import matplotlib.pyplot as plt
import time
from multiprocessing import Pool
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 20
```

# 1.Implement the map, plot the evolution of x

```python
def bif_map(x_0, r, n=200):
    x = np.array([x_0])
    for _ in np.arange(n - 1):
        x_1 = r * x[-1] * (1 - x[-1])
        x.append(x_1)
    return x
```

```python
def bifurcation(x_0, r, m=0, n=200):
    assert isinstance(r, (list, tuple, set, np.ndarray))
    assert m <= n

    arr_dict = {}
    for r_i in r:
        x_ri = bif_map(x_0, r_i, n)
        arr_dict[r_i] = x_ri[-int(m):]
    return arr_dict
```

```python
x_0 = np.random.rand()
r = np.linspace(0, 4, 5)
n = 100
```

```python
# Evolution of x
arr_dict = bifurcation(x_0=x_0, r=r, n=n)

for key, v in zip(arr_dict.keys(), arr_dict.values()):
    plt.plot(range(n), v, label=f'r = {round(key, 1)}')

plt.legend()
plt.suptitle('Evolution of x for different constant  r')
plt.xlabel('Iterations after n = {}'.format(n))
plt.ylabel('$x_{n}$')
```
