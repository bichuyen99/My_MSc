<h2> Problem 1.0 (Matrix norms) </h2>

For any norm $\|\cdot\|$ in finite dimentional space $\mathbb{C}^{N}$ we define dual norm $\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$, where ${\cdot}^{\star}$ denotes transpose complex-conjugate vector.
<h3>1.0.1. Prove that $\|\cdot\|^{'}$ is a norm</h3> 

With any norm $ \left\|\cdot\right\| $ in finite dimentional space $ \mathbb{C} ^ N $, satisfy certain properties: <br>
- $ \left\|x\right\| \geq 0 $, if $ \left\|x\right\| = 0 $ then $ x = 0 $ 
- $ \left\|\alpha x\right\| = |\alpha|\left\|x\right\| $ 
- $ \left\|x+y\right\| \leq \left\|x\right\| + \left\|y\right\| $ <br><br>
To be a norm $ \left\|\cdot\right\|' $ should satisfy above properties. <br>
Note that in finite dimentional space $ \mathbb{C} ^ N: y^{\star}x = x\cdot y = \sum\limits_{i=1}^{n}x_{i}\bar y_{i} $ <br>

+ $x = 0: \|x\|^{'} = 0$
+ $x \neq 0:$ <br>
We have $\hat{z} := \frac{x}{\|x\|}, \|\hat{z}\|=1$. Compare with $\|y\|=1$, so we can replace $y$ with $\hat{z}:$
$$ \|x\|' = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\} \\ = \sup\left\{\left|\hat{z}^{\star} x\right| \right\} = \sup\left\{\left|x\cdot\hat{z}\right| \right\} \geq \frac{\|x\|^{2}}{\|x\|} \geq 0 $$
Therefore, $ \left\|x\right\|' \geq 0 $ for all $ x \in \mathbb{C} ^ N $, if $ \left\|x\right\|' = 0 $ then $ x = 0 ~~~~~ (1)$ <br>
Next to the second property: <br>
$$ \|\alpha x\|' = sup\{\alpha\cdot|y^*x|\} = |\alpha|\cdot sup{\{|y^*x|\}} = |\alpha|\cdot\|x\|' ~~~~~~ (2) $$ 
Lastly, <br>
$$ \left\|x+z\right\|' = sup\{|y^*(x+z)|\} = sup\{|y^*x+y^*z|\} \leq sup\{|y^*x|\}+ sup\{|y^*z|\} = \left\|x\right\|' + \left\|z\right\|' $$ \
$$ \Rightarrow \left\|x+z\right\|' \leq \left\|x\right\|' + \left\|z\right\|' ~~~~~~~~~ (3) $$  
From (1), (2) and (3) we can conclude that $ \|\cdot\|' $ is a norm.

<h3>1.0.2. Find dual norms for Chebyshev norm $\left\|x\right\|_{\infty} = \max\limits_{i=1,\dots, N}|x_{i}|$ and $l_1$ norm $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$. </h3>

Let $y = (y_{1},...,y_{n}) \in \mathbb{C} ^ N$ with $\|y\|_{q} \leq 1 $ <br>
As we know: $$y^{\star}x = x\cdot y = \sum\limits_{i=1}^{n}x_{i}\bar y_{i} \leq \sum\limits_{i=1}^{n}|x_{i}\bar y_{i}| =\|xy\|_{1} $$
By HÃ¶lder's inequality, let $ 1 \leq p,q \leq \infty $ with $ \frac{1}{p} + \frac{1}{q} = 1$, we have <br>
$$\|xy\|_{1} \leq \|x\|_{p}\|y\|_{q}$$
Therefore, 
$$y^{\star}x = x\cdot y \leq \|xy\|_{1} \leq \|x\|_{p}\|y\|_{q} \leq \|x\|_{p} $$ 
So we can predict that $\sup\left\{\left|{y}^{\star} x\right|, \|y\|_{q} \leq 1 \right\} = \|x\|_{p}$ <br>
To improve that, find $z \in \mathbb{C} ^ N$ with $\|z\|_q = 1$ and $\sum_{i=1}^{n}x_iz_i=\|x\|_p.$ <br>
Because $\|z\|_{q} = 1 $, choose $z := \frac{y}{\|y\|_{q}} ~ (\|y\|_q>0$ with $x \neq 0)$ then $$ \sum_{i=1}^{n} x_i z_i = \sum_{i=1}^{n} x_{i}\frac{y_i}{\|y\|_q} = \frac{1}{\|y\|_q}\sum_{i=1}^{n} x_i y_i  ~~~~~~ (4)$$
Let $y_i := \text{csgn} (x_i) |x_i|^{p-1},~ i = 1,...,n$; note that $|x_{i}| = x_{i}csgn(x_i)$. <br> 
$$\sum_{i=1}^{n} x_i y_i = \sum_{i=1}^{n} x_{i}\text{csgn} (x_i)|x_i|^{p-1} =\sum_{i=1}^{n} |x_i|^p = \|x\|_p^p ~~~~~~ (5)$$
Beside that,
$$\|y\|_{q}^q = \sum_{i=1}^{n} |y_i|^q = \sum_{i=1}^{n} \big|csgn(x_i)|x_i|^{p-1}\big|^q = \sum_{i=1}^{n} |x_i|^{(p-1)q} = \sum_{i=1}^{n} |x_i|^p = \|x\|_p^p $$ 
Here, $(p-1)q=p$ from $\frac{1}{p}+\frac{1}{q}=1$. 
$$\Rightarrow \|y\|_q = {(\|y\|_{q}^q)}^{\frac{1}{q}}={(\|x\|_{p}^p)}^{\frac{1}{q}} =\|x\|_{p}^{\frac{p}{q}} ~~~~~~ (6)$$
From (4), (5) and (6), we have: <br>
$$ \sum_{i=1}^{n} x_i z_i = \frac{1}{\|y\|_q}\sum_{i=1}^{n} x_i y_i = \frac{1}{\|x\|_{p}^{\frac{p}{q}}}\|x\|_p^p=\|x\|_p^{p-\frac{p}{q}}= \|x\|_p$$
Thus we found $\|z\|_q=1$ so that $\sum_{i=1}^{n}x_iz_i=\|x\|_p$, i.e. $$ \|x\|' = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|_q=1\right\} = \|x\|_{p}$$
Conclude that $\left\|\cdot\right\|_p$ is a dual norm of $\left\|\cdot\right\|_q$ when $ \frac{1}{p} + \frac{1}{q} = 1$ for $ 0 \leq p,q \leq \infty $

Therefore, <br>
- dual norm for Chebyshev norm $ \left\|x\right\|_\infty = \max_{i = 1,...,N}|x_i| $ is $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$; <br>
- dual norm for $l_1$ norm $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$ is $ \left\|x\right\|_\infty = \max_{i = 1,...,N}|x_i| $ <br> 
Because of $\frac{1}{\infty} + \frac{1}{1} = 1$

<h3>1.0.4. Find dual norm for weighted norm $\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$ with positive-definite matrix $A$ and $x\in\mathbb{R}^{N}$ </h3>

With positive-definite matrix $A$ and $x\in\mathbb{R}^{N}$, exist 2 positive-definite matrix $A^\frac{1}{2}$ and $A^\frac{-1}{2}$. <br>
Choose $ x = A^{\frac{-1}{2}}y $, we have:
$$\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$$ 
$$\Leftrightarrow \left\|A^{\frac{-1}{2}}y\right\|_{A} = \left((A^{\frac{-1}{2}}y)^T A A^{\frac{-1}{2}}y\right)^{1/2}$$ 
$$\Leftrightarrow \left\|y\right\|_{A} = \left(y^TAy\right)^{1/2}$$ 
The dual norm for weighted norm $\left\|x\right\|_{A}$ is:
$$\|x\|^{'}_{A} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\} = \sup\left\{\frac{\left|y^{\star} x\right|}{\left\|y\right\|}:y\neq 0\right\} =\sup\frac{\left|y^{\star} x\right|}{\sqrt{y^{T}Ay}} $$ 
Then multiply both numerator and denominator with $ A^{\frac{-1}{2}}$ so
$$\|x\|^{'}_{A} =\sup\frac{\left|(A^{\frac{-1}{2}}y)^{\star} x\right|}{\sqrt{y^{T}A^{\frac{-1}{2}}AA^{\frac{-1}{2}}y}} =\sup\frac{\left|(A^{\frac{-1}{2}}y)^{\star} x\right|}{\sqrt{y^{T}y}}= \sup\frac{\left|(A^{\frac{-1}{2}}y)^{\star} x\right|}{{\|y\|}_{2}} =$$
$$  = \sup\frac{\left|y^{\star} (A^{\frac{-1}{2}}x)\right|}{{\|y\|}_{2}} = \frac{{{\|A^{\frac{-1}{2}}x\|}^2}_2}{\|A^{\frac{-1}{2}}x\|_2} = \|A^{\frac{-1}{2}}x\|_2 = A^{\frac{-1}{2}}\sqrt{x^{T}x} = \sqrt{x^{T}A^{-1}x}  = \|x\|_{A^{-1}} $$
Thus $\|x\|_{A^{-1}} $ is dual norm for weighted norm $\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$

```python
import matplotlib.pyplot as plt
import pandas as pd
```

We have in our posession a database with molecules. Each molecule has five atoms `A`, `B`, `C`, `D`, `E` located in the same plane $x, y$.

The coordinates of the molecules was collected by an automatic system that processed "pictures" from the electronic microscope. As a result, some of the samples differs by unknown rotation in $2D$ space. Your task is to normalize this database to find distinct molecules.

```python
data = pd.read_csv("https://raw.githubusercontent.com/oseledets/nla2022/main/hw1/molecules.csv", index_col=0)
data.head()
```

```python
molecules = data.to_numpy().reshape(len(data), 5, 2)
molecules.shape
```

```python
def draw_molecule(ax, molecule):
    ax.scatter(*molecule.T, color="black", s=50)
    for i in range(1, molecule.shape[0]):
        ax.plot(*molecule[[0, i], :].T, color="black")
    ax.axis("off")
    ax.axis("equal")
```

```python
k = 5
fig, ax = plt.subplots(k, k, figsize=(2*k, 2*k))
ax = ax.reshape(-1,)
for i in range(k**2):
    draw_molecule(ax[i], molecules[i])
```

+ Prove that orthogonal transformations preserve $l_2$ norm, i.e., $\left\|x\right\|_2 = \left\|Ux\right\|_2$, where $U$ is an orthogonal matrix.


$$\left\|Ux\right\|_2 = \langle Ux,Ux \rangle_2 = \langle x,U^{\top}Ux\rangle_2 = \langle x,x\rangle_2 = \left\|x\right\|_2$$


+ Use the result from the previous point to construct an algorithm that distinguishes different molecules. Implement it and run on the database. Produce a list of distinct molecules and visualise them.

```python
import torch.nn as nn
import torch
import numpy as np
from scipy.stats import ortho_group
```

```python
#Create a random orthogonal U
U = ortho_group.rvs(dim=5) 
np.dot(np.transpose(U), U)
```

As proved, orthogonal transformations preserve  $l_2$  norm. So we find different molecules by comparing their $l_2$ norm. (up to the 14th decimal places)

```python
distinct_norm = []
distinct_mol = []

for i in range(molecules.shape[0]):
  current = molecules[i]
  dot = np.dot(U, current)
  norm = round(np.linalg.norm(dot, 2),15)

  if norm not in distinct_norm:
    distinct_norm.append(norm)
    distinct_mol.append(current)

len(distinct_mol)
```

```python
def kabsch(P, Q):
  # translation: center the points
  centr_P = np.mean(P, axis=0)
  centr_Q = np.mean(Q, axis=0)
  P -= centr_P
  Q -= centr_Q

  # covariance matrix: 
  H = np.dot(np.transpose(P), Q)

  # svd decomp
  U, S, Vt = np.linalg.svd(H)
  d = (np.linalg.det(U) * np.linalg.det(Vt)) < 0.0
  if d:
    S[-1] = -S[-1]
    U[:, -1] = -U[:, -1]

  # create rotation matrix R
  R = np.dot(U, Vt)
  return R

def rmsd(V, W):
    D = len(V[0])
    N = len(V)
    rmsd = 0.0
    for v, w in zip(V, W):
        rmsd += sum([(v[i] - w[i])**2.0 for i in range(D)])
    return np.sqrt(rmsd/N)

def kabsch_rmsd(P, Q):
  R = kabsch(P, Q)
  P_rotate = np.dot(P, R)
  error = rmsd(P_rotate, Q)
  return error
```

```python
distinct_mol2 = [molecules[0]]

for i in range(1, molecules.shape[0]):
  same = False
  current = molecules[i]

  for existing_mol in distinct_mol2:
    error = kabsch_rmsd(current, existing_mol)
    if error <= 1e-14:
      same = True
  
  if same == False:
    distinct_mol2.append(current)


len(distinct_mol2)
```

## Brief intro

TheÂ mainÂ componentÂ ofÂ [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html)Â isÂ GeneratorÂ network.Â GeneratorÂ isÂ suchÂ modelÂ thatÂ takesÂ asÂ inputÂ randomÂ vectorÂ andÂ producesÂ anÂ image.Â InÂ thisÂ task,Â weÂ willÂ workÂ withÂ theÂ alreadyÂ pre-trainedÂ generatorÂ $G$.Â HereÂ areÂ aÂ fewÂ thingsÂ youÂ needÂ toÂ knowÂ aboutÂ thisÂ model:



1.Â Â Â **ModelÂ input:**Â batchÂ ofÂ randomÂ vectorsÂ $z\simÂ N(0,1)$Â withÂ normallyÂ distributedÂ valuesÂ ofÂ sizeÂ **(Batch_Size,100)**.Â Here,Â Batch_SizeÂ isÂ a number ofÂ imagesÂ weÂ wantÂ toÂ generate
2.Â Â Â **ModelÂ output:**Â $\mathrm{images}Â =Â G(z)$Â -Â batchÂ ofÂ imagesÂ withÂ sizeÂ **(Batch_Size,3,64,64)**.Â HereÂ **3**Â -Â definesÂ numberÂ ofÂ imageÂ channelsÂ (3Â forÂ RGBÂ images),Â **64,Â 64**Â -Â widthÂ andÂ height.
3.Â Â Â **ModelÂ structure:**Â generatorÂ isÂ aÂ setÂ ofÂ sequentiallyÂ apppliedÂ functions.Â EachÂ functionÂ inputÂ isÂ theÂ outputÂ ofÂ the previousÂ one.Â InÂ otherÂ words,Â $G(z)Â =Â f_N(f_{N-1}...f_1(f_0(z)))$.Â $N$Â -Â definesÂ numberÂ ofÂ functionsÂ inÂ generator.



### Disentanglement

TheÂ goodÂ propertyÂ ofÂ GANÂ isÂ thatÂ generatorÂ inputÂ spaceÂ hasÂ aÂ setÂ ofÂ meaningfulÂ directionsÂ -Â movingÂ inputÂ vectorÂ alongÂ thisÂ directionsÂ produceÂ interpretableÂ changesÂ inÂ image.Â ForÂ example,Â weÂ canÂ changeÂ aÂ Pose,Â Age,Â ExpressionÂ orÂ evenÂ addÂ eyeglasses.Â YouÂ canÂ readÂ moreÂ aboutÂ itÂ onÂ [resource](https://genforce.github.io/interfacegan/).Â TheÂ mainÂ questionÂ isÂ howÂ toÂ findÂ suchÂ goodÂ directions.Â ThisÂ problemÂ isÂ calledÂ disentanglement.

![](https://github.com/genforce/interfacegan/blob/master/docs/assets/teaser.jpg?raw=true)

## Problem discription 

WeÂ areÂ goingÂ toÂ findÂ meaningfulÂ directionsÂ usingÂ SVDÂ appliedÂ overÂ theÂ Jacobian matrixÂ ofÂ someÂ hiddenÂ outputÂ withÂ respectÂ toÂ modelÂ input. **i-th**Â hiddenÂ outputÂ ofÂ generatorÂ modelÂ is defined by formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$,Â whereÂ $iÂ <Â N$.Â So,Â theÂ taskÂ couldÂ beÂ splittedÂ onÂ fourÂ mainÂ parts:

1.Â **(4Â pts)**Â Â Â WriteÂ theÂ functionÂ **Get_Hidden_Output**,Â whichÂ reproducesÂ formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$
2.Â **(3Â pts)**Â Â Â CalculateÂ jacobian matrixÂ $J$Â ofÂ generatorÂ $G$Â forÂ severalÂ **i**Â values.Â ForÂ eachÂ ofÂ them:

*Â Â Â CalculateÂ theÂ jacobian matrixÂ withÂ respectÂ toÂ $z$:Â $\frac{\partialÂ G(z)}{\partialÂ z}$
*Â Â Â ReshapeÂ jacobian matrixÂ toÂ haveÂ theÂ sizeÂ **(Batch_Size * Hidden_Output_Size,100)**

3.Â **(3Â pts)**Â Â ApplyÂ SVDÂ toÂ jacobianÂ matrixÂ ($JÂ =Â UÂ @Â SÂ @Â V.T$)Â fromÂ previousÂ step

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
      if index == 0:
        inp = inp[:,:,None,None]
      return self.main[index](inp)
    def get_number_of_layers(self):
      return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```
